```python
import pandas as pd
import xml.etree.ElementTree as ET

# Load and parse the XML content
tree = ET.parse("/kaggle/input/xml-training/sheet1.xml")
root = tree.getroot()

# Define namespace
ns = {'ns': 'http://schemas.openxmlformats.org/spreadsheetml/2006/main'}

# Extract rows
rows = root.findall('.//ns:sheetData/ns:row', ns)

# Extract all data
data = []
max_col = 0
for row in rows:
    cells = row.findall('ns:c', ns)
    row_data = {}
    for cell in cells:
        cell_ref = cell.attrib.get('r')
        cell_col = ''.join(filter(str.isalpha, cell_ref))
        cell_value_elem = cell.find('ns:v', ns)
        cell_value = cell_value_elem.text if cell_value_elem is not None else ''
        row_data[cell_col] = cell_value
    data.append(row_data)
    max_col = max(max_col, len(row_data))

# Convert column letters to ordered list
all_cols = sorted({col for row in data for col in row})
df = pd.DataFrame([{col: row.get(col, '') for col in all_cols} for row in data])

# Save to Excel and CSV
excel_path = "./complete_output.xlsx"
csv_path = "./complete_output.csv"
df.to_excel(excel_path, index=False)
df.to_csv(csv_path, index=False)

excel_path, csv_path
```
